## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5)) # width - height

# 3. Plot data
ax.plot(x, y);

# 4. Customize plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel='y-axis');

# 5. Save and show
# fig.savefig('sample-plot.png')
```

```python
# Make a plot from dictionary
nut_butter_prices = {'Almond butter': 10,
                     'Peanut butter': 8,
                     'Cashew butter': 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Nut's butter store",
       ylabel='Price ($)');
```

```python
fig, ax = plt.subplots()
ax.bar(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
```

```python
# Make some data for histogram and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))

# Plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(x);
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
# # Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(x);
```

### Which one should you use? (pyplot vs matplotlib OO method)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use OO method

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
# OO method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
ax.set(xlim=[45, 100]);
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target']);

# Customize the plot
ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol');

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
over_50.head()
```

```python
# Subplots of chol, age, thalach
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,10),sharex=False)

# Add data to ax0
scatter0 = ax[0].scatter(x=over_50['age'],
                         y=over_50['chol'],
                         c=over_50['target']);
ax[0].set(title='Heart Disease and Cholesterol Levels',
          xlabel='Age',
          ylabel='Cholesterol');
ax[0].legend(*scatter0.legend_elements(), title='Target');
ax[0].axhline(over_50['chol'].mean(),linestyle='--');

# Add data to ax1
scatter1 = ax[1].scatter(x=over_50['age'],
                         y=over_50['thalach'],
                         c=over_50['target']);
ax[1].set(title='Heart Disease and Max Heart Rate',
          xlabel='Age',
          ylabel='Max Heart Rate');
ax[1].legend(*scatter1.legend_elements(), title='Target');
ax[1].axhline(over_50['thalach'].mean(),
              linestyle='--');

# Add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```

```python
plt.style.use('seaborn-whitegrid')

## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter');

# Customize the plot
ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol');

# Add a legend
ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
           linestyle='--');
```

```python
# Customizing the y and x axis limitation

# Subplots of chol, age, thalach
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10,10),sharex=True)

# Add data to ax0
scatter0 = ax[0].scatter(x=over_50['age'],
                         y=over_50['chol'],
                         c=over_50['target'],
                         cmap='winter');
ax[0].set(title='Heart Disease and Cholesterol Levels',
          ylabel='Cholesterol');
# Change the x0 axis limit
ax[0].set_xlim([50, 80])
ax[0].legend(*scatter0.legend_elements(), title='Target');
ax[0].axhline(over_50['chol'].mean(),linestyle='--');

# Add data to ax1
scatter1 = ax[1].scatter(x=over_50['age'],
                         y=over_50['thalach'],
                         c=over_50['target'],
                         cmap='winter');
# Change the x1 axis limit
ax[1].set_xlim([50, 80])
ax[1].set(title='Heart Disease and Max Heart Rate',
          xlabel='Age',
          ylabel='Max Heart Rate');
ax[1].legend(*scatter1.legend_elements(), title='Target');
ax[1].axhline(over_50['thalach'].mean(),
              linestyle='--');

# Add a title to the figure
fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold');
```
